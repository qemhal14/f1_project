```python
import numpy as np
import pandas as pd
import json
from urllib.request import urlopen
```

# 1. Driver Standings

```python
url = 'http://ergast.com/api/f1/current/driverStandings.json'
response = urlopen(url)
raw_data = response.read().decode('utf-8')

data = json.loads(raw_data)
        
df_standings = pd.json_normalize(data['MRData']['StandingsTable']['StandingsLists'][0]['DriverStandings'])
df_standings.head()
```

```python
df_standings.info()
```

```python
df_standings["Constructors"] = df_standings["Constructors"].apply(lambda x:x[0]["constructorId"])
df_standings["Constructors"]
```

```python
df_standings["Constructors"] = df_standings["Constructors"].replace("alpine", "BWT Alpine F1 Team")
df_standings["Constructors"] = df_standings["Constructors"].replace("haas", "MoneyGram Haas F1 Team")
df_standings["Constructors"] = df_standings["Constructors"].replace("mercedes", "Mercedes-AMG PETRONAS F1 Team")
df_standings["Constructors"] = df_standings["Constructors"].replace("williams", "Williams Racing")
df_standings["Constructors"] = df_standings["Constructors"].replace("aston_martin", "Aston Martin Aramco F1 Team")
df_standings["Constructors"] = df_standings["Constructors"].replace("sauber", "Stake F1 Team Kick Sauber")
df_standings["Constructors"] = df_standings["Constructors"].replace("rb", "Visa Cash App RB Formula One Team")
df_standings["Constructors"] = df_standings["Constructors"].replace("ferrari", "Scuderia Ferrari")
df_standings["Constructors"] = df_standings["Constructors"].replace("mclaren", "McLaren Formula 1 Team")
df_standings["Constructors"] = df_standings["Constructors"].replace("red_bull", "Oracle Red Bull Racing")
df_standings["Constructors"]
```

```python
df_standings.head()
```

```python
df_standings["full_name"] = df_standings['Driver.givenName'] + ' ' + df_standings['Driver.familyName']
df_standings["full_name"]
```

```python
df_standings["Driver.permanentNumber"] = df_standings["Driver.permanentNumber"].replace("33", "1")
df_standings.head()
```

# 2. Constructors Standings

```python
import xml.etree.ElementTree as ET
import pandas as pd
from urllib.request import urlopen

# Load the data
url = 'http://ergast.com/api/f1/current/constructorStandings'
response = urlopen(url)
raw_data = response.read().decode('utf-8')

root = ET.fromstring(raw_data)

namespace = {'ns': 'http://ergast.com/mrd/1.5'}
standings_list = []

for standing in root.findall('.//ns:ConstructorStanding', namespaces=namespace):
    position = standing.get('position')
    points = standing.get('points')
    wins = standing.get('wins')
    constructor = standing.find('ns:Constructor/ns:Name', namespaces=namespace).text
    nationality = standing.find('ns:Constructor/ns:Nationality', namespaces=namespace).text
    
    standings_list.append({
        'Position': position,
        'Points': points,
        'Wins': wins,
        'Constructor': constructor,
        'Nationality': nationality
    })

# Convert to DataFrame
df_standings_c = pd.DataFrame(standings_list)
df_standings_c
```

# 3. Mclaren Constructor Standings

```python
url = 'https://ergast.com/api/f1/constructors/mclaren/constructorstandings?limit=30&offset=30'
response = urlopen(url)
raw_data = response.read().decode('utf-8')
print(raw_data) 
```

```python
root = ET.fromstring(raw_data)

namespace = {'ns': 'http://ergast.com/mrd/1.5'}
standings_list = []

for standing in root.findall('.//ns:ConstructorStanding', namespaces=namespace):
    position = standing.get('position')
    points = standing.get('points')
    wins = standing.get('wins')

    constructor_element = standing.find('ns:Constructor', namespaces=namespace)
    if constructor_element is not None:
        constructor = constructor_element.find('ns:Name', namespaces=namespace).text
        nationality = constructor_element.find('ns:Nationality', namespaces=namespace).text
    
    standings_list.append({
        'Position': position,
        'Points': points,
        'Wins': wins,
        'Constructor': constructor,
        'Nationality': nationality
        })

# Convert to DataFrame
mclaren_standings = pd.DataFrame(standings_list)
mclaren_standings
```

```python
mclaren_standings.loc[24] = ['2', '404', '3', 'McLaren', 'British']
mclaren_standings
```

```python
mclaren_standings["Year"] = np.arange(2000, 2025)
mclaren_standings
```

```python
mclaren_standings.rename(columns={col: f"{col}_mc" for col in mclaren_standings.columns if col != "Year"}, inplace=True)
mclaren_standings
```
